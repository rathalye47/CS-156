#### Load libraries and set random number generator seed

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.exceptions import ConvergenceWarning
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import cross_val_score
```

```python
np.random.seed(42)
```

#### Code the solution

#### Load the MNIST dataset

```python
digits = datasets.load_digits()
X = digits.data
Y = digits.target
```

#### Normalize the MNIST dataset

```python
X = X.astype('float32') / 255
```

#### Split the data into training and test sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0, stratify=Y)
```

#### Train the 1st MLP model, collect stratified 5-fold cross-validation accuracies for each fold, and compute test set prediction accuracy

```python
all_cross_val_scores = [] # List to store cross-validation accuracies for each fold for all models.
model_names = [] # List to store model names.
all_test_scores = [] # List to store test set accuracies for all models.

model1 = MLPClassifier(random_state=1, max_iter=10000).fit(X_train, Y_train) # Trains the model.
cross_vals1 = list(cross_val_score(model1, X_train, Y_train, cv=5)) # Collects cross-validation accuracies for each fold.
test1 = model1.score(X_test, Y_test) # Computes test set accuracy.

# Adds cross-validation accuracy of each fold to a list.
for fold_score in cross_vals1:
  all_cross_val_scores.append(fold_score)

# Adds the model name to a list.
for i in range(5):
  model_names.append("Model 1")

# Adds test set accuracy to a list.
all_test_scores.append(test1)
```

#### Train the 2nd MLP model, collect stratified 5-fold cross-validation accuracies for each fold, and compute test set prediction accuracy

```python
model2 = MLPClassifier(hidden_layer_sizes=(400, 150, 50), max_iter=10000, activation='relu', random_state=1).fit(X_train, Y_train) # Trains the model.
cross_vals2 = list(cross_val_score(model2, X_train, Y_train, cv=5)) # Collects cross-validation accuracies for each fold.
test2 = model2.score(X_test, Y_test) # Computes test set accuracy.

# Adds cross-validation accuracy of each fold to a list.
for fold_score in cross_vals2:
  all_cross_val_scores.append(fold_score)

# Adds the model name to a list.
for i in range(5):
  model_names.append("Model 2")

# Adds test set accuracy to a list.
all_test_scores.append(test2)
```

#### Train the 3rd MLP model, collect stratified 5-fold cross-validation accuracies for each fold, and compute test set prediction accuracy

```python
model3 = MLPClassifier(hidden_layer_sizes=(64, 32, 8), max_iter=10000, activation='relu', random_state=1).fit(X_train, Y_train) # Trains the model.
cross_vals3 = list(cross_val_score(model3, X_train, Y_train, cv=5)) # Collects cross-validation accuracies for each fold.
test3 = model3.score(X_test, Y_test) # Computes test set accuracy.

# Adds cross-validation accuracy of each fold to a list.
for fold_score in cross_vals3:
  all_cross_val_scores.append(fold_score)

# Adds the model name to a list.
for i in range(5):
  model_names.append("Model 3")

# Adds test set accuracy to a list.
all_test_scores.append(test3)
```

#### Train the 4th MLP model, collect stratified 5-fold cross-validation accuracies for each fold, and compute test set prediction accuracy

```python
model4 = MLPClassifier(hidden_layer_sizes=(32, 16), max_iter=10000, activation='relu', random_state=1).fit(X_train, Y_train) # Trains the model.
cross_vals4 = list(cross_val_score(model4, X_train, Y_train, cv=5)) # Collects cross-validation accuracies for each fold.
test4 = model4.score(X_test, Y_test) # Computes test set accuracy.

# Adds cross-validation accuracy of each fold to a list.
for fold_score in cross_vals4:
  all_cross_val_scores.append(fold_score)

# Adds the model name to a list.
for i in range(5):
  model_names.append("Model 4")

# Adds test set accuracy to a list.
all_test_scores.append(test4)
```

#### Train the 5th MLP model, collect stratified 5-fold cross-validation accuracies for each fold, and compute test set prediction accuracy

```python
model5 = MLPClassifier(hidden_layer_sizes=(120, 64, 16), max_iter=10000, activation='relu', random_state=1).fit(X_train, Y_train) # Trains the model.
cross_vals5 = list(cross_val_score(model5, X_train, Y_train, cv=5)) # Collects cross-validation accuracies for each fold.
test5 = model5.score(X_test, Y_test) # Computes test set accuracy.

# Adds cross-validation accuracy of each fold to a list.
for fold_score in cross_vals5:
  all_cross_val_scores.append(fold_score)

# Adds the model name to a list.
for i in range(5):
  model_names.append("Model 5")

# Adds test set accuracy to a list.
all_test_scores.append(test5)
```

#### Train the 6th MLP model, collect stratified 5-fold cross-validation accuracies for each fold, and compute test set prediction accuracy

```python
model6 = MLPClassifier(hidden_layer_sizes=(320, 120, 32), max_iter=10000, activation='relu', random_state=1).fit(X_train, Y_train) # Trains the model.
cross_vals6 = list(cross_val_score(model6, X_train, Y_train, cv=5)) # Collects cross-validation accuracies for each fold.
test6 = model6.score(X_test, Y_test) # Computes test set accuracy.

# Adds cross-validation accuracy of each fold to a list.
for fold_score in cross_vals6:
  all_cross_val_scores.append(fold_score)

# Adds the model name to a list.
for i in range(5):
  model_names.append("Model 6")

# Adds test set accuracy to a list.
all_test_scores.append(test6)
```

#### Plot the results from the cross-validation and test sets for all the models

```python
# Creates a dataframe to store cross-validation accuracies of each fold for all models.
cross_val_results_df = pd.DataFrame({"Model": model_names, "Accuracy": all_cross_val_scores})

# Creates a dataframe to store test set accuracies for all models.
model_names_2 = ["Model 1", "Model 2", "Model 3", "Model 4", "Model 5", "Model 6"]
test_results_df = pd.DataFrame({"Model": model_names_2, "Accuracy": all_test_scores})

# Creates a violin plot to show cross-validation and test set accuracies for all models.
plt.figure(figsize=(11,7))
plt.title("Cross-Validation and Test Set Accuracies across all Models", pad=10)
sns.violinplot(x="Model", y="Accuracy", data=cross_val_results_df, inner="quartile")
sns.swarmplot(x="Model", y="Accuracy", data=test_results_df, color="pink", size=10);
```
