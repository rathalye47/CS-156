#### Code the solution

```python
display(Math(r' f(x) = 5x^3 - 20x + 2 '))
```

```python
def f(x):
    return 5*pow(x, 3) - 20*x + 2
```

```python
# Derivative of f(x).
def f_prime(x):
    return 15*pow(x, 2) - 20
```

```python
init = np.random.randint(-10, 10)

# Function to implement the gradient descent algorithm.
def gradient_descent(x, lr, iterations):
    for i in range(iterations):
        x = x - lr * f_prime(x)
    return(x)

solution = gradient_descent(init, 0.01, 1000)
print("The global minimum is at x = "+str(round(solution, 3)))
```
